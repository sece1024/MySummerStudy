# 使用正则表达式抓取网页内容

```python
import requests
import re

myUrl="http://www.people.com.cn/"
myContent=requests.get(myUrl).content.decode('gb2312')
print(myContent)
```

```python
#<a href="http://liuyan.people.com.cn/threads/content?tid=7929181" target="_blank">网友:防止老旧小区改造猫腻 关键公开透明</a>
myPatten="<a href=\"(.*)\" target=\"_blank\">(.*)</a>"
myList=re.findall(myPatten,myContent)
myList
```

```python
for item in myList:
    oneNews={}
    oneNews['title']=item[1]
    oneNews['href']=item[0]
    print(oneNews)
```
